# Analyzing Income Data

## Agricultural Income

```python
#Importing Income Database (previously created above)
incomedf = pd.read_csv('Income.csv')

#For now, replace all NaN's by 0. In future, may wish to do this on variable-by-variable basis
incomedf.fillna(0, inplace=True)

#Calculating Total Ag Income
incomedf['TotalAgIncome'] = incomedf['incann'] + incomedf['incbeef'] + incomedf['inccalf'] + incomedf['incchicken'] + incomedf['incdonkey'] + incomedf['incmilk'] + incomedf['incper'] + incomedf['incsheepgoat'] + incomedf['inchf']

#Remove observations with negative values (e.g. didn't answer) and set range for any outliers to be removed
incomedf_positive = incomedf.loc[(incomedf['TotalAgIncome'] > 0) & (incomedf['TotalAgIncome'] < 5000000)]

incomedf_positive

```

## Off-Farm Income

```python
#Calculating Off-Farm Income (but not including government transfers)
incomedf_positive['TotalNonFarmIncome'] = incomedf_positive['inccompany'] + incomedf_positive['incdaily'] + incomedf_positive['incoff'] + incomedf_positive['incother'] + incomedf_positive['incother_rent'] + incomedf_positive['incrental_lot'] + incomedf_positive['incretail'] + incomedf_positive['incsalary']

incomedf_positive['incsalary'].describe()
```

## Government Transfer Income

```python
#Calculating Government Transfer Income

incomedf_positive['TotalGovIncome'] = incomedf_positive['incbolsafam'] + incomedf_positive['incpension'] + incomedf_positive['incbolsaescola']

incomedf_positive['TotalGovIncome'].describe()
```

## Cleaning Data

```python
#Adjusting income levels for inflation using CPI
cpi_values = pd.read_excel('CPI1990-2021.xlsx')

incomedf_positive['CPIValue'] = 100

incomedf_positive.loc[incomedf.year ==2000, 'CPIValue'] = cpi_values['2000 [YR2000]'][0]
incomedf_positive.loc[incomedf.year ==2005, 'CPIValue'] = cpi_values['2005 [YR2005]'][0]
incomedf_positive.loc[incomedf.year ==2009, 'CPIValue'] = cpi_values['2009 [YR2009]'][0]
incomedf_positive.loc[incomedf.year ==2015, 'CPIValue'] = cpi_values['2015 [YR2015]'][0]
incomedf_positive.loc[incomedf.year ==2019, 'CPIValue'] = cpi_values['2019 [YR2019]'][0]

income_categories =['incann', 'incbeef', 'inccalf', 'incchicken', 'incdonkey', 'incmilk', 'incper', 'incsheepgoat', 'inchf', 'inccompany', 'incdaily', 'incoff', 'incother', 'incother_rent', 'incrental_lot', 'incretail', 'incsalary','incbolsafam', 'incpension', 'incbolsaescola', 'TotalGovIncome',  'TotalNonFarmIncome', 'TotalAgIncome']

for c in income_categories:
    incomedf_positive['Norm' + c + '2010'] = incomedf_positive[c] / (0.01 * incomedf_positive['CPIValue'])

#Create Total Household Income Variable
incomedf_positive['NormTotalIncome2010'] = incomedf_positive['NormTotalAgIncome2010'] + incomedf_positive['NormTotalNonFarmIncome2010'] + incomedf_positive['NormTotalGovIncome2010']
```

## Plots

### Bar Plot of Mean Household Income for each year in survey, broken down by income source

```python
#Calculate and plot average income by hh over panel years
year = incomedf_positive.groupby(['year'])['year'].mean()

#Ag Income
ag_income = incomedf_positive.groupby(['year'])['NormTotalAgIncome2010'].mean()
beef_income = incomedf_positive.groupby(['year'])['Normincbeef2010'].mean()
milk_income = incomedf_positive.groupby(['year'])['Normincmilk2010'].mean()
calf_income = incomedf_positive.groupby(['year'])['Norminccalf2010'].mean()
crop_income = incomedf_positive.groupby(['year'])['Normincann2010'].mean()

#Non-Ag Income
nonfarm_income = incomedf_positive.groupby(['year'])['NormTotalNonFarmIncome2010'].mean()
daily_income = incomedf_positive.groupby(['year'])['Normincdaily2010'].mean()
salary_income = incomedf_positive.groupby(['year'])['Normincsalary2010'].mean()

#Government Transfers
gov_income = incomedf_positive.groupby(['year'])['NormTotalGovIncome2010'].mean()


ag_income_std = incomedf_positive.groupby(['year'])['NormTotalAgIncome2010'].std()

ag_income_low_percentile = np.zeros(len(year))
ag_income_high_percentile = np.zeros(len(year))


for i in range(len(year)):
    ag_income_low_percentile[i] = np.percentile(incomedf_positive.loc[incomedf_positive['year'] == year.iloc[i]]['NormTotalAgIncome2010'], 5)
    ag_income_high_percentile[i] = np.percentile(incomedf_positive.loc[incomedf_positive['year'] == year.iloc[i]]['NormTotalAgIncome2010'], 95)

annual_count = incomedf_positive.groupby('year')['NormTotalAgIncome2010'].count()



f, ax1 = plt.subplots(figsize=(15,10))

x = incomedf_positive['year']

ax1.bar(np.arange(len(year)), ag_income, alpha = 0.25, label='Other Ag Income', color='chartreuse')
ax1.bar(np.arange(len(year)), beef_income, alpha=0.75, label = 'Beef', color='forestgreen')
ax1.bar(np.arange(len(year)), milk_income, bottom=beef_income, alpha=0.75, label = 'Milk', color='lime')
ax1.bar(np.arange(len(year)), calf_income, bottom=beef_income + milk_income, alpha=0.75, label = 'Calf', color='mediumseagreen')

ax1.bar(np.arange(len(year)), nonfarm_income, alpha = 0.25, bottom = ag_income, label = 'Other NonFarm Income', color='darkviolet')
ax1.bar(np.arange(len(year)), daily_income, alpha = 0.75, bottom = ag_income, label = 'Daily Wage Income', color='violet')
ax1.bar(np.arange(len(year)), salary_income, alpha = 0.75, bottom = ag_income + daily_income, label = 'Salary Income', color='thistle')

ax1.bar(np.arange(len(year)), gov_income, alpha = 0.25, bottom = ag_income + nonfarm_income, label = 'Government Income', color='goldenrod')



#ax1.bar(np.arange(len(year)), crop_income, bottom=beef_income + milk_income + chicken_income, alpha=0.75, label = 'Crops')

#ax1.errorbar(np.arange(len(year)), annual_income, yerr=[annual_income_16percentile, annual_income_84percentile], fmt="o")
ax1.set_xticks(ticks = np.arange(len(year)), labels = ['1996', '2000', '2005', '2009', '2019'], fontsize=18)
ax1.set_ylabel('Average Household Income (2010 Rs)', fontsize=20)
ax1.set_ylim([0,70000])
plt.yticks(size=18)

ax2 = ax1.twinx()
ax2.scatter(np.arange(len(year)), annual_count, marker = 'x', color='red', linewidth=10, label='Observations')
ax2.set_ylabel('Number of Households', fontsize=20)
plt.yticks(size=18)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left', fontsize=18)


#plt.show()
plt.savefig('AnnualIncomebySource.png', bbox_to_inches = 'tight', facecolor='w')
```
